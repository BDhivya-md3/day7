```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("/content/Social_Network_Ads.csv")
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1]
print(x)
print(y)
```

```python
#splitting the dataset into training and test sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.25)
print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
print(x_train)
print(x_test) 
```

```python
#logistic Regression Implementation
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
#predict the test set results
y_pred = classifier.predict(x_test)
```

```python
#predict the new entry
print(classifier.predict(sc.transform([[20,850000]])))
```

```python
cm = confusion_matrix(y_pred,y_test)
print(cm)
accuracy_score(y_pred,y_test)
```

```python
#visualizing training set result
from matplotlib.colors import ListedColormap
x_set,y_set = sc.inverse_transform(x_train),y_train
x11,x22 = np.meshgrid(np.arange(start = x_set[:,0].min() -0,stop = x_set[:,0].max() +0,step=0.5),
                    np.arange(start = x_set[:,1].min() -0,stop = x_set[:,1].max() + 1,step = 0.25))

plt.contourf(x11,x22,classifier.predict(sc.transform(np.array([x11.ravel(),x22.ravel()]).T)).reshape(x11.shape),cmap = ListedColormap(('red','green')))

plt.xlim(x11.min(), x11.max())
plt.ylim(x22.min(), x22.max())

for i,j in enumerate(np.unique(y_set)):
  plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], c = ListedColormap(("red", "green"))(i), label = j)
plt.title('logistic regression')
plt.xlabel('age')
plt.ylabel('salary')
plt.legend()
plt.show()
```
